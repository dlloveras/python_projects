```python
import numpy as np
import matplotlib.pyplot as plt
phi1=  np.radians(np.linspace(0,360/4, 500))
phi2=  np.radians(np.linspace(360/4,360/2, 500))
phi3=  np.radians(np.linspace(360/2,360*3/4, 500))
phi4=  np.radians(np.linspace(360*3/4,360, 500))
phi5=  np.radians(np.linspace(360,360*5/4, 500))
phi6=  np.radians(np.linspace(360*5/4,360*6/4, 500))
phi7=  np.radians(np.linspace(360*6/4,360*7/4, 500))
phi8=  np.radians(np.linspace(360*7/4,360*8/4, 500))

phi11=  np.radians(np.linspace(0,360/4, 500)) +360*8/4
phi21=  np.radians(np.linspace(360/4,360/2, 500))+360*8/4
phi31=  np.radians(np.linspace(360/2,360*3/4, 500))+360*8/4
phi41=  np.radians(np.linspace(360*3/4,360, 500))+360*8/4
phi51=  np.radians(np.linspace(360,360*5/4, 500))+360*8/4
phi61=  np.radians(np.linspace(360*5/4,360*6/4, 500))+360*8/4
phi71=  np.radians(np.linspace(360*6/4,360*7/4, 500))+360*8/4
phi81=  np.radians(np.linspace(360*7/4,360*8/4, 500))+360*8/4
params=[10,0,0,0.9,0.5]

def func(params,phi):
    F=params[3]*(1-(params[4]*np.cos(phi))+3*params[4]*np.sin(phi/4))
    x = params[0]*F*np.cos(phi)#+params[1]                  #aca solo phi es la variable independiente
    y = params[0]*np.sin(phi)
    return x,y
```

```python

x,y=func(params,phi1)
plt.plot(x,y,'r')
x,y=func(params,phi2)
plt.plot(x,y,'g')
x,y=func(params,phi3)
plt.plot(x,y,'b')
x,y=func(params,phi4)
plt.plot(x,y,'black')
x,y=func(params,phi5)
plt.plot(x,y,'r')
x,y=func(params,phi5)
plt.plot(x,y,'g')
x,y=func(params,phi6)
plt.plot(x,y,'b')
x,y=func(params,phi7)
plt.plot(x,y,'black')

x,y=func(params,phi11)
plt.plot(x,y,'r')
x,y=func(params,phi21)
plt.plot(x,y,'g')
x,y=func(params,phi31)
plt.plot(x,y,'b')
x,y=func(params,phi41)
plt.plot(x,y,'black')
x,y=func(params,phi51)
plt.plot(x,y,'r')
x,y=func(params,phi51)
plt.plot(x,y,'g')
x,y=func(params,phi61)
plt.plot(x,y,'b')
x,y=func(params,phi71)
plt.plot(x,y,'black')

plt.show()
```
